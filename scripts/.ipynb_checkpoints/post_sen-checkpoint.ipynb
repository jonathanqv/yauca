$ Sobol!sti$

```python
stiPath = os.path.join(path,F"{ws.lower()}.sobol.sti.csv")
sti = pd.read_csv(stiPath,index_col=0)
cols = sti.columns
pars = [i.split('-')[1].split(':')[0].replace('_inst','') for i in cols]
sti.columns=pars
```

```python
fig,ax=plt.subplots(figsize=(12,3))
im = ax.imshow(sti.values,aspect='auto',cmap='cividis')
cbar = fig.colorbar(im, ax=ax, extend='both')
cbar.ax.set_ylabel('Total sensitivity index')
ax.set_ylabel('Model runs')
ax.grid(color='white',alpha=0.5)
ax.set_xticks(range(len(sti.columns)))
ax.set_xticklabels(pars, rotation=90)
ax.set_title('Yauca watershed')
plt.show()
fig.savefig('figs_rep/sensitivity_yauca.png', transparent=True,bbox_inches='tight', dpi=300)
```

```python
hs = ['gwflow_coef','gwstor_min']
pars_lowsen = [i for i in pars if i not in hs]
sti_lowsen = sti.loc[:,pars_lowsen]

fig,ax=plt.subplots(figsize=(12,3))
sti_lowsen.loc['phi',:].plot(ax=ax)
ax.set_ylabel('Model runs')
ax.grid(color='white',alpha=0.5)
ax.set_xticks(range(len(sti_lowsen.columns)))
ax.set_xticklabels(pars_lowsen, rotation=90)
ax.set_title('Yauca watershed')
plt.show()
```

$ Sobol!obs$

```python
obsPath = os.path.join(path,F'{ws.lower()}.sobol.obs.csv')
obs = pd.read_csv(obsPath,index_col=0)
obs.columns = ['failed_flag'] + [i.split(':')[-1] for i in obs.columns if 'oname' in i]
obs = obs[obs['failed_flag']==0]
obs = obs[[i for i in obs.columns if i != 'failed_flag']]
obs = obs.T
obs.index= pd.to_datetime(obs.index)
obs = obs.sort_index()
obscols = obs.columns
```

```python
obs_streamgage = pd.concat([obs,streamgage],axis=1).dropna()
```

```python
statDf = pd.DataFrame(index=obscols,columns=['NSE','RMSE'])
for idx,row in statDf.iterrows():
    NSE,RMSE=stats(obs_streamgage.loc[:,'Flows'],obs_streamgage.loc[:,idx],silent=True)
    statDf.loc[idx,'NSE'] = NSE
    statDf.loc[idx,'RMSE'] = RMSE
statDf.NSE.max()
```
