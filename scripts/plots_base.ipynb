```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from variables import *
from helpers import stats
from matplotlib import rc
from Figures import ReportFigures
rf = ReportFigures()
rf.set_style(width='double', height='tall')
rc('text', usetex=True)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
```

```python
outFlow = pd.read_csv(os.path.join(outPath,'nseg_seg_outflow.csv'),index_col=0,parse_dates=True)
```

```python
obsFlow = pd.read_csv(os.path.join(obsPath,'flow_yauca.csv'),index_col=0,parse_dates=True)
obsFlow[obsFlow == -999] = np.nan
```

```python
obsFlow.dropna().shape[0]/pd.date_range('1/1/2014','12/31/2016').shape[0]
```

```python
NSE, RMSE = stats(obsFlow['Flows'],outFlow[' 1'],'best_09262022')
```

```python
fig,ax=plt.subplots(figsize=(12,6))
outFlow[' 1'].plot(ax=ax,color='#F96666')
obsFlow.plot(ax=ax,color='#647E68')
ax.set_title('Simulated vs observed flows for the Yauca \nstreamgage station in the Yauca watershed')
ax.set_ylabel('Streamflow (cfs)')
ax.grid(which ='both',alpha=0.5,color='black')
ax.text('2016',3000,r'\bf{NSE = '+F"{NSE:.2f}"+'}', fontsize=16, fontweight='bold') # 
handles, labels = ax.get_legend_handles_labels()
rf.legend(ax, handles, ['simulated','observed'], bbox_to_anchor=(.95, -0.22),fontsize=10)
fig.savefig('figs_rep/streamflow_yauca.png', transparent=True,bbox_inches='tight', dpi=300)
```
