```python
basePath = os.path.join(auxPath,'Yauca.gpkg')
base = gpd.read_file(basePath)
```

```python
tmaxPath = os.path.join(outPath,'nhru_tmaxf.csv')
tmax = pd.read_csv(tmaxPath,index_col=0,parse_dates=True)

tminPath = os.path.join(outPath,'nhru_tminf.csv')
tmin = pd.read_csv(tmaxPath,index_col=0,parse_dates=True)

ppPath = os.path.join(outPath,'nhru_hru_ppt.csv')
pp = pd.read_csv(ppPath,index_col=0,parse_dates=True)
```

```python
date = '2014-01-04'
base['tmax'] = tmax.loc[date].values
base['tmin'] = tmin.loc[date].values
base['pp'] = pp.loc[date].values* 25.4

print(base['pp'].min(),base['pp'].max())
print(base['tmax'].min(),base['tmax'].max())
print(base['tmin'].min(),base['tmin'].max())

base.plot(column='pp',cmap='viridis',legend=True)
```
